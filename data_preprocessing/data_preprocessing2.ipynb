#### Converts Any String into Real Python Object

```python
import ast
df_merged["genres"] = df_merged["genres"].map(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df_merged["cast"] = df_merged["cast"].map(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df_merged["crew"] = df_merged["crew"].map(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

#### Get the Genres List

```python
def make_genresList(x):
    gen = []

    for i in x:
        name = i.get('name')

        if name == "Science Fiction":
            gen.append("Sci-Fi")
        else:
            gen.append(name)

    return np.NaN if not gen else " ".join(gen)

df_merged['genres_list'] = df_merged['genres'].map(make_genresList)
```

#### Extract Actors and Director

```python
def get_actors(x, index):
    casts = [i.get('name') for i in x]
    if len(casts) <= index:
        return np.NaN
    return casts[index]

# def get_directors(x):
#     directors =[i.get('name') for i in x if i.get('job') == "Directors"]

#     return np.NaN if not directors else " ".join(directors)

def get_directors(x):
    if not isinstance(x, list):
        return np.NaN
        
    directors = [i.get('name') for i in x if i.get('job') == 'Director']
    
    return np.NaN if not directors else " ".join(directors)



df_merged["actor_1_name"] = df_merged["cast"].map(lambda x: get_actors(x, 0))
df_merged["actor_2_name"] = df_merged["cast"].map(lambda x: get_actors(x, 1))
df_merged["actor_3_name"] = df_merged["cast"].map(lambda x: get_actors(x, 2))

df_merged["director_name"] = df_merged["crew"].map(lambda x: get_directors(x)) 
```
