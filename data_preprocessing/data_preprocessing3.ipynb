#### Extract the 2018 Released Movies From Wikipidia

```python
import pandas as pd
import requests

years = range(2018, 2026)

headers = {
    "User-Agent": "Mozilla/5.0"
}

all_movies = []

for year in years:
    url = f"https://en.wikipedia.org/wiki/List_of_American_films_of_{year}"
    
    try:
        response = requests.get(url, headers=headers)
        tables = pd.read_html(response.text, header=0)
        
        # Wikipedia movie pages usually store movie tables from index 2 onward
        movie_tables = tables[2:]
        
        df_year = pd.concat(movie_tables, ignore_index=True)
        df_year["year"] = year   # Add year column
        
        all_movies.append(df_year)
        
        print(f"{year} loaded successfully")
        
    except Exception as e:
        print(f"Error loading {year}: {e}")

# Combine all years
movies_df = pd.concat(all_movies, ignore_index=True)

movies_df.head()
```

```python
from tmdbv3api import TMDb
from dotenv import load_dotenv
import json
import requests
import os

load_dotenv()

api_key = os.getenv("TMDB_API_KEY")
tmdb = TMDb()
tmdb.api_key = api_key
```

```python
from tmdbv3api import Movie

tmdb_movie = Movie()

def get_genres(title):
    try:
        result = tmdb_movie.search(title)

        if len(result) == 0:
            return np.NaN

        movie_id = result[0].id
        movie_details = tmdb_movie.details(movie_id)

        if movie_details.genres:
            genres = [g['name'] for g in movie_details.genres]
            return " ".join(genres)

        return np.NaN

    except:
        return np.NaN
```

```python
new_df['genres'] = new_df['Title'].astype(str).map(get_genres)
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
new_df["director_name"] = new_df["Cast and crew"].map(lambda x: get_director(str(x)))
```

```python
def extract_movie_people(x):
    result = {
        "actor_1_name": np.nan,
        "actor_2_name": np.nan,
        "actor_3_name": np.nan
    }

    try:
        actors_part = x.split("screenplay); ")[-1]  
        actors_list = [a.strip() for a in actors_part.split(", ")]
        if len(actors_list) >= 1:
            result["actor_1_name"] = actors_list[0]
        if len(actors_list) >= 2:
            result["actor_2_name"] = actors_list[1]
        if len(actors_list) >= 3:
            result["actor_3_name"] = actors_list[2]
    except:
        pass  

    return result

df_all = new_df.copy()
people_df = df_all["Cast and crew"].map(extract_movie_people).apply(pd.Series)

df_all.loc[:, ["actor_1_name", "actor_2_name", "actor_3_name"]] = people_df
```

```python
df_all = df_all.rename(columns={"Title": "movie_title"})
```

```python
new_all_df = df_all[["movie_title", "director_name", "actor_1_name", "actor_2_name", "actor_3_name", "genres"]]
```

```python
new_all_df["actor_1_name"] = new_all_df["actor_1_name"].replace(np.nan, "unknown")
new_all_df["actor_2_name"] = new_all_df["actor_2_name"].replace(np.nan, "unknown")
new_all_df["actor_3_name"] = new_all_df["actor_3_name"].replace(np.nan, "unknown")
```

```python
new_all_df["movie_title"] = new_all_df["movie_title"].str.lower()
```
